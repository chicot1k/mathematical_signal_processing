```python
import glob
import librosa

# Заполняем список классов
classes = []
for filepath in glob.iglob('/Users/sasha/Desktop/ESC-10/*'):
    print(filepath)
    classes.append(filepath)

print(classes)

# Инициализируем пустой список для данных
data = []
label_number = 0

# Проходим по каждому классу
for i in classes:
    print("the class = " + i + ", the label = " + str(label_number))
    # Ищем файлы в текущем классе
    for j in glob.iglob(i + '/*'):
        print(f"Loading file {j}")  # Добавим отладочное сообщение
        y, s = librosa.load(j, sr=8000)  # Может down-sample до 8000 Гц?
        data.append([y, label_number])
        
    label_number += 1
```

Из аудио-сигналов извлечём признаки с помощью ОПФ. Спектрограмму векторизуем.
Длительность ограничим 437675 отсчётами.

```python

sigLength = 437675
from tqdm import tqdm 
stft = np.zeros((400, sigLength))

for i in tqdm(range(0,400)):
    freq = librosa.amplitude_to_db(np.abs(librosa.stft(audio_data[i])), ref=np.max)
    # ref - параметр за масштабирование
    freq=freq.reshape(-1)
    
    if freq.shape[0]<sigLength:
        freq = np.array(list(freq) + [0]*(sigLength-freq.shape[0]))
        stft[i] = freq        
    else:
        stft[i] = freq[:sigLength]
        
stft = np.array(stft) 
stft.shape
```

Далее требуется подготовить данные для обучения свёрточной нейронной сети в PyTorch

```python
import torch
import torch.nn as nn
```

Классы закодирем с помощью one-hot encoding.

```python
def encode_labels(labels, num_of_classes): 
    num = len(labels)
    target = torch.zeros(num, num_of_classes).float()
    for i, l in enumerate(labels):
        target[i, l] = 1
    return target
```

```python
target = encode_labels(labels, label_number)
print(target.shape)
print(target[0])
```

Признаки сигналов (то есть ОПФ)  нормализуем с помощью MinMaxScaler() из пакета sklearn (приведение значений в интервал от 0 до 1).

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
scaler.fit(stft)

normalized_stft = scaler.transform(stft)

print(np.amax(stft))
print(np.amax(normalized_stft))
```

```python
model = nn.Sequential(
    nn.Conv2d(1,16,kernel_size=3,stride=2,padding=1),
    nn.ReLU(),
    nn.MaxPool2d(2,2),
    
    nn.Conv2d(16,32,kernel_size=3,stride=2,padding=1),
    nn.ReLU(),
    nn.MaxPool2d(2,2),

    nn.Conv2d(32,64,kernel_size=3,stride=2,padding=1),
    nn.ReLU(),
    nn.MaxPool2d(2,2),
    
    nn.Flatten(),
    nn.Linear(64*112,64),
    nn.Linear(64,10)
)
```

```python
if torch.cuda.is_available():
    print("cuda")
    device=torch.device('cuda:0')
else:
    print("cpu")
    device=torch.device('cpu')

model.to(device)
```

```python
import torch.optim as optim
from torch.optim import Adam

learning_rate = 5e-3

loss_fn = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

train_losses = []
valid_losses = []
```

```python
import torch
from torch.utils.data import DataLoader, TensorDataset
import numpy as np
import tqdm
```

```python
# Преобразуем данные в тензоры PyTorch
X_train = torch.tensor(X_train, dtype=torch.float32)
X_test = torch.tensor(X_test, dtype=torch.float32)
Y_train = torch.tensor(Y_train, dtype=torch.long)
Y_test = torch.tensor(Y_test, dtype=torch.long)
```

```python
train_dataset = TensorDataset(X_train, Y_train)
test_dataset = TensorDataset(X_test, Y_test)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
```

```python
import tqdm
epochs = 30
for epoch in tqdm.tqdm(range(1,epochs+1)):
    model.train()
    batch_losses=[]

    for i, data in enumerate(train_loader):
        x, y = data
        optimizer.zero_grad()
        x = x.to(device, dtype=torch.float32)
        y = y.to(device, dtype=torch.float32)
        y_hat = model(x)
        loss = loss_fn(y_hat, y)
        loss.backward()
        batch_losses.append(loss.item())
        optimizer.step()
    train_losses.append(batch_losses)
    # print(f'Epoch - {epoch} Train-Loss : {np.mean(train_losses[-1])}')
    model.eval()
    batch_losses=[]
    trace_y = []
    trace_yhat = []
    for i, data in enumerate(test_loader):
        x, y = data
        x = x.to(device, dtype=torch.float32)
        y = y.to(device, dtype=torch.float32)
        y_hat = model(x)
        loss = loss_fn(y_hat, y)
        trace_y.append(y.cpu().detach().numpy())
        trace_yhat.append(y_hat.cpu().detach().numpy())      
        batch_losses.append(loss.item())
    valid_losses.append(batch_losses)
    trace_y = np.concatenate(trace_y)
    trace_yhat = np.concatenate(trace_yhat)
    accuracy = np.mean(trace_yhat.argmax(axis=1)==trace_y.argmax(axis=1))
    print(f'Epoch - {epoch}\n Accuracy : {accuracy}')
```
